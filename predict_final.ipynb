## Prediction
#### Target = BARTHAG
Power Rating (Chance of beating the average Division I basketball team).

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Assuming your DataFrame with the historical data is named 'df_historical'
# Assuming your DataFrame with the 68 teams' data for the 2024 season is named 'df_2024_teams'

# Define the features and target variable for training
features = ['BADJ EM', 'BADJ O', 'BADJ D', 'EFG%', 'EFG%D', 'FTR', 'FTRD', 'TOV%', 'TOV%D', 'OREB%', 'DREB%', 'OP OREB%', 'OP DREB%', 'RAW T', '2PT%', '2PT%D', '3PT%', '3PT%D', 'BLK%', 'BLKED%', 'AST%', 'OP AST%', '2PTR', '3PTR', '2PTRD', '3PTRD', 'BADJ T', 'AVG HGT', 'EFF HGT', 'EXP', 'TALENT', 'FT%', 'OP FT%', 'PPPO', 'PPPD', 'ELITE SOS', 'SEED']
target = 'BARTHAG'

# Prepare the training data
X_train = stats_df[features]
y_train = stats_df[target]

# Train a new random forest regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict the BARTHAG for each team in the 2024 season
predicted_barthag = {}
for _, team in df_2024_teams.iterrows():
    team_name = team['TEAM']
    team_features = [team[col] for col in features]
    barthag = rf_model.predict([team_features])[0]
    predicted_barthag[team_name] = barthag

# Sort the teams by their predicted BARTHAG
ranked_teams_barthag = sorted(predicted_barthag.items(), key=lambda x: x[1], reverse=True)
```

#### Target = BADJ EM
Estimates how many points a team would outscore the average Division I basketball team by over the course of 100

```python
# Define the features and target variable for training
features = ['BADJ O', 'BADJ D', 'EFG%', 'EFG%D', 'FTR', 'FTRD', 'TOV%', 'TOV%D', 'OREB%', 'DREB%', 'OP OREB%', 'OP DREB%', 'RAW T', '2PT%', '2PT%D', '3PT%', '3PT%D', 'BLK%', 'BLKED%', 'AST%', 'OP AST%', '2PTR', '3PTR', '2PTRD', '3PTRD', 'BADJ T', 'AVG HGT', 'EFF HGT', 'EXP', 'TALENT', 'FT%', 'OP FT%', 'PPPO', 'PPPD', 'ELITE SOS', 'SEED']
target = 'BADJ EM'

# Prepare the training data
X_train = stats_df[features]
y_train = stats_df[target]

# Train a new random forest regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict the BADJ EM for each team in the 2024 season
predicted_badj_em = {}
for _, team in df_2024_teams.iterrows():
    team_name = team['TEAM']
    team_features = [team[col] for col in features]
    badj_em = rf_model.predict([team_features])[0]
    predicted_badj_em[team_name] = badj_em

# Sort the teams by their predicted BADJ EM
ranked_teams_badj_em = sorted(predicted_badj_em.items(), key=lambda x: x[1], reverse=True)
```

```python
import random

def simulate_game(team1, team2, predicted_badj_em):
    team1_badj_em = predicted_badj_em[team1]
    team2_badj_em = predicted_badj_em[team2]
    
    # Calculate the probability of team1 winning based on the difference in BADJ EM
    prob_team1_win = 1 / (1 + 10 ** ((team2_badj_em - team1_badj_em) / 15))
    
    # Simulate the game outcome based on the probability
    if random.random() < prob_team1_win:
        return team1
    else:
        return team2
```

```python
def simulate_round(matchups, predicted_badj_em):
    winners = []
    for team1, team2 in matchups:
        winner = simulate_game(team1, team2, predicted_badj_em)
        winners.append(winner)
    return winners
```

## Tournament Simulaton using variable BADJ EM as target
+ BADJ EM - Estimates how many points a team would outscore the average Division I basketball team by over the course of 100
+ source: Kaggle

```python
def simulate_game(team1, team2, df_2024_teams):
    team1_badj_em = df_2024_teams.loc[df_2024_teams['TEAM'] == team1, 'BADJ EM'].values[0]
    team2_badj_em = df_2024_teams.loc[df_2024_teams['TEAM'] == team2, 'BADJ EM'].values[0]
    
    # Calculate the probability of team1 winning based on the difference in BADJ EM
    prob_team1_win = 1 / (1 + 10 ** ((team2_badj_em - team1_badj_em) / 15))
    
    # Simulate the game outcome based on the probability
    if random.random() < prob_team1_win:
        return team1, prob_team1_win
    else:
        return team2, 1 - prob_team1_win

def simulate_round(matchups, df_2024_teams):
    winners = []
    convictions = []
    for team1, team2 in matchups:
        winner, conviction = simulate_game(team1, team2, df_2024_teams)
        winners.append(winner)
        convictions.append(conviction)
    return winners, convictions

# Simulate the tournament rounds
first_round_winners, first_round_convictions = simulate_round(first_round_matchups, df_2024_teams)
print("First Round Results:")
for i, (team1, team2, winner, conviction) in enumerate(zip(first_round_df['Team_1'], first_round_df['Team_2'], first_round_winners, first_round_convictions), 1):
    print(f"Game {i}: {team1} vs {team2} | Winner: {winner} (Conviction: {conviction:.2%})")

second_round_winners, second_round_convictions = simulate_round(second_round_matchups, df_2024_teams)
print("\nSecond Round Results:")
for i, (team1, team2, winner, conviction) in enumerate(zip(second_round_df['Team_1'], second_round_df['Team_2'], second_round_winners, second_round_convictions), 1):
    print(f"Game {i}: {team1} vs {team2} | Winner: {winner} (Conviction: {conviction:.2%})")

sweet_16_winners, sweet_16_convictions = simulate_round(sweet_16_matchups, df_2024_teams)
print("\nSweet 16 Results:")
for i, (team1, team2, winner, conviction) in enumerate(zip(sweet_16_df['Team_1'], sweet_16_df['Team_2'], sweet_16_winners, sweet_16_convictions), 1):
    print(f"Game {i}: {team1} vs {team2} | Winner: {winner} (Conviction: {conviction:.2%})")

elite_8_winners, elite_8_convictions = simulate_round(elite_8_matchups, df_2024_teams)
print("\nElite 8 Results:")
for i, (team1, team2, winner, conviction) in enumerate(zip(elite_8_df['Team_1'], elite_8_df['Team_2'], elite_8_winners, elite_8_convictions), 1):
    print(f"Game {i}: {team1} vs {team2} | Winner: {winner} (Conviction: {conviction:.2%})")

final_4_winners, final_4_convictions = simulate_round(final_4_matchups, df_2024_teams)
print("\nFinal 4 Results:")
for i, (team1, team2, winner, conviction) in enumerate(zip(final_4_df['Team_1'], final_4_df['Team_2'], final_4_winners, final_4_convictions), 1):
    print(f"Game {i}: {team1} vs {team2} | Winner: {winner} (Conviction: {conviction:.2%})")

print("\nChampionship Result:")
champion, conviction = simulate_game(championship_df.loc[0, 'Team_1'], championship_df.loc[0, 'Team_2'], df_2024_teams)
print(f"{championship_df.loc[0, 'Team_1']} vs {championship_df.loc[0, 'Team_2']} | Winner: {champion} (Conviction: {conviction:.2%})")
```
